```python
import pandas as pd
# Load the dataset
data = pd.read_csv(r'mubawab_listings_eda.csv', dtype=object)
```

```python
# One-Hot Encode categorical columns
cleaned_data_encoded = pd.get_dummies(data, columns=['Type', 'Quartier', 'Status', 'Property_State'])

# Display the first few rows of the cleaned and encoded dataset
cleaned_data_encoded.head()
```

```python
Y = cleaned_data_encoded['Price']
```

```python
X = cleaned_data_encoded.drop(columns=['Price'])
```

```python
X = X.drop(columns=['Price_m2'])
X.head()
```
